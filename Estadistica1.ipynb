```python
import numpy as np
import scipy as sc
from scipy.integrate import odeint 
import random
import pylab
import statistics
from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.optimize import curve_fit
from scipy.special import factorial
from scipy.stats import poisson
```

```python
#Defines the N number of steps 
n = 500
#Defines the Number of times the Random Walk will be ran
Samples = 5000

#Lists where the Information about position will be stored
Xf = []
Yf = []
```

```python
#Random Walk Function, takes N steps variable and returns Final Position at X and Y

def walkAlt(n):
    angle = 0
    X = 0
    Y = 0
    #Inner Loop, with each tick will use the random module to chose a direction between 0 and 2pi 
    for i in range(n):
        angle = random.randrange(0,20)
        X = X + np.cos(((angle / 10 )*np.pi))
        Y = Y + np.sin(((angle / 10 )*np.pi))

    return X, Y 
```

```python
#Repetition Loop, will repeat the random Walk for the set number of sanples
for i in range(Samples):
    X, Y = walkAlt(n)
    Xf.append(X)
    Yf.append(Y) 
```

```python
#random walk plot
plt.scatter (Xf, Yf, c="hotpink", alpha=0.2, s=3)
```

```python
#Distances List
R=[]
#Calculation of the distance R^2
for i in range(len(Yf)):
    r= ( (Xf[i])**2 + (Yf[i])**2 )
    R.append(r)
```

```python
# best fit of data
(mu, sigma) = norm.fit(Xf)

# the histogram of the data
n, bins, patches = plt.hist(Xf, 40, density=1, facecolor='Pink', alpha=0.75)

# add a 'best fit' line
y = norm.pdf( bins, mu, sigma)
l = plt.plot(bins, y, 'Slateblue', linewidth=2)
Mode = statistics.mode(Xf)

#plot
plt.xlabel('X Distance')
plt.title(("Mean=" + str(round(mu,3)) + " mode= " + str(round(Mode,3)) + " Sigma= " + str(round(sigma,3)) ))
plt.grid(True)

plt.show()
```

```python
# best fit of data
(mu, sigma) = norm.fit(Yf)

# the histogram of the data
n, bins, patches = plt.hist(Yf, 40, density=1, facecolor='Skyblue', alpha=0.75)

# add a 'best fit' line
y = norm.pdf( bins, mu, sigma)
l = plt.plot(bins, y, 'Hotpink', linewidth=2)
Mode = statistics.mode(Yf)

#plot
plt.xlabel('Y Distance')
plt.title(("Mean=" + str(round(mu,3)) + " mode= " + str(round(Mode,3)) + " Sigma= " + str(round(sigma,3)) ))
plt.grid(True)

plt.show()
```
